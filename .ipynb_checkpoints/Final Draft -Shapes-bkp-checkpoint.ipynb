```python
import numpy as np
import os
import PIL
import PIL.Image
import tensorflow as tf
from sklearn.impute import SimpleImputer
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.image as mpimg
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.utils.vis_utils import plot_model
from tensorflow.keras.optimizers import RMSprop
import matplotlib.pyplot as plt
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras import regularizers
import random
from PIL import Image
```

```python
# tf.keras.utils.set_random_seed(1)
np.random.seed(1)
```

```python
path ="./trafficsigns_dataset/"
lst = []
for root, dirs, files in os.walk(path):
    for file in files:
        if(file.endswith(".png")):
#             print(root)
            root1=root.split("/")[-1].split("\\")
            filepath=path+root1[0]+'/'+root1[1]+'/'+file
            lst.append([root1[0], root1[1], file, filepath])
dfdata = pd.DataFrame(lst, columns=['shape', 'sign', 'imageName','filePath'])
```

```python
fig=plt.figure(figsize=(15,15))
imglist=dfdata['filePath'].values.tolist()
random_paths=random.sample(imglist,5)
flag=0
for i in random_paths:
    fig.add_subplot(4,4,flag+1)
    img = mpimg.imread(i)
    plt.axis("off")
    imgplot = plt.imshow(img)
    
    plt.show()
    flag=flag+1
    
```

```python
model_1_df=pd.DataFrame()
model_1_df['Path']=dfdata['filePath']
model_1_df['Shape']=dfdata['shape']
```

```python
model_1_df
```

```python
trainX,valX,trainY,valY=train_test_split(model_1_df["Path"],model_1_df["Shape"],test_size=0.2,shuffle=True,random_state=10)
```

```python
trainY
trainY.value_counts()
```

```python
valY.value_counts()
```

```python
train_df=pd.DataFrame()
train_df["Paths"]=trainX
train_df["Shapes"]=trainY
val_df=pd.DataFrame()
val_df["Paths"]=valX
val_df["Shapes"]=valY
```

```python
val_df.shape
```

```python
#imagedatagenerators
```

```python
idg_train=ImageDataGenerator(rescale=1/255.0,data_format='channels_last')
idg_val=ImageDataGenerator(rescale=1/255.0,data_format='channels_last')
```

```python
train_gen=idg_train.flow_from_dataframe(dataframe=train_df,
                                        batch_size=32,
                                        seed=10,
                                        class_mode='categorical',
                                        target_size=(28, 28),
                                        x_col='Paths',
                                        y_col='Shapes',
                                        color_mode = 'grayscale')
val_gen=idg_val.flow_from_dataframe(dataframe=val_df,
                                       batch_size=32,
                                       seed=10,
                                       class_mode='categorical',
                                       target_size=(28, 28),
                                       x_col='Paths',
                                       y_col='Shapes',
                                       color_mode = 'grayscale'
                                      )
```

```python
train_gen
```

```python
#baseline--tiny
```

```python
baseline_model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28,1)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(5,activation='softmax')

])
```

```python
baseline_model.summary()
```

```python
tf.keras.utils.plot_model(baseline_model, show_shapes=True)
```

```python
pip install pydot
```

```python
baseline_model.compile(optimizer='SGD',loss='categorical_crossentropy',
              metrics=['categorical_accuracy'])
```

```python
fitted_baseline=baseline_model.fit(train_gen,validation_data=val_gen,batch_size=32,epochs=50,shuffle=False)
```

```python
plt.title('shapes acc')
plt.plot(fitted_baseline.history['categorical_accuracy'],label='tr_acc')
plt.plot(fitted_baseline.history['val_categorical_accuracy'],label='val_acc')
plt.legend(loc='best')
```

```python
plt.plot(fitted_baseline.history['loss'],label='tr_loss')
plt.plot(fitted_baseline.history['val_loss'],label='val_loss')
plt.legend(loc='best')
```

```python
weights = compute_class_weight("balanced",np.unique(train_gen.classes),train_gen.classes)
```

```python
weights_dict={}
for i in range(weights.shape[0]):
    weights_dict[np.unique(train_gen.classes)[i]]=weights[i]
```

```python
weights_dict
```

```python
#training weighted model
```

```python
baseline_model_weighted = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28,1)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(5,activation='softmax')
])
baseline_model_weighted.compile(optimizer='SGD',loss='categorical_crossentropy',
              metrics=['categorical_accuracy'])
fitted_baseline_weighted=baseline_model_weighted.fit(train_gen,validation_data=val_gen,batch_size=32,epochs=50,class_weight=weights_dict,shuffle=False)
```

```python
plt.title('shapes acc')
plt.plot(fitted_baseline_weighted.history['categorical_accuracy'],label='tr_acc')
plt.plot(fitted_baseline_weighted.history['val_categorical_accuracy'],label='val_acc')
plt.legend(loc='best')
```

```python
new_idg_train=ImageDataGenerator(rescale=1/255.0,
                                 data_format='channels_last',
                                 rotation_range=95, #type data cannot do
                                 #width_shift_range=0.2,
                                 #height_shift_range=0.2,
                                 #shear_range=0.2,
                                 zoom_range=0.2,
                                 fill_mode='nearest',
                                 horizontal_flip=True, #type data cannot do
                                 vertical_flip=True)#type data cannot do
new_idg_val=ImageDataGenerator(rescale=1/255.0,
                                 data_format='channels_last',
                                 rotation_range=95,
                                 #width_shift_range=0.2,
                                 #height_shift_range=0.2,
                                 #shear_range=0.2,
                                 zoom_range=0.2,
                                 fill_mode='nearest',
                                 horizontal_flip=True,
                                 vertical_flip=True)
new_train_gen=new_idg_train.flow_from_dataframe(dataframe=train_df,
                                        batch_size=32,
                                        seed=10,
                                        class_mode='categorical',
                                        target_size=(28, 28),
                                        x_col='Paths',
                                        y_col='Shapes',
                                        color_mode = 'grayscale')
new_val_gen=new_idg_val.flow_from_dataframe(dataframe=val_df,
                                       batch_size=32,
                                       seed=10,
                                       class_mode='categorical',
                                       target_size=(28, 28),
                                       x_col='Paths',
                                       y_col='Shapes',
                                       color_mode = 'grayscale'
                                      )
```

```python
#Using best model from above baseline model with new image data generators (Augmentation) and max pooling layer
```

```python
best_aug = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(5,(3,3),activation='relu',input_shape=(28,28,1)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    tf.keras.layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    tf.keras.layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    tf.keras.layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),

    tf.keras.layers.Dense(5,activation='softmax')
    
])
best_aug.compile(optimizer='adam',loss='categorical_crossentropy',
              metrics=['categorical_accuracy'])
best_aug_fit=best_aug.fit(new_train_gen,validation_data=new_val_gen,batch_size=32,epochs=100,class_weight=weights_dict,shuffle=False)
```

```python
plt.plot(best_aug_fit.history['categorical_accuracy'],label='tr_acc')
plt.plot(best_aug_fit.history['val_categorical_accuracy'],label='val_acc')
plt.legend(loc='best')
```

```python
path ="./test_data_final_scaled/"
lst = []
for root, dirs, files in os.walk(path):
    for file in files:
        if(file.endswith(".png")):
#             print(root)
            root1=root.split("/")[-1].split("\\")
            filepath=path+root1[0]+'/'+root1[1]+'/'+file
            lst.append([root1[0], root1[1], file, filepath])
dfdata_test = pd.DataFrame(lst, columns=['shape', 'sign', 'imageName','filePath'])
```

```python
new_idg_train_shape=ImageDataGenerator(rescale=1/255.0,
                                 data_format='channels_last',
                                 rotation_range=95, #type data cannot do
                                 #width_shift_range=0.2,
                                 #height_shift_range=0.2,
                                 #shear_range=0.2,
                                 zoom_range=0.2,
                                 fill_mode='nearest',
                                 horizontal_flip=True, #type data cannot do
                                 vertical_flip=True)#type data cannot do

test_gen_shape_new = new_idg_train_shape.flow_from_dataframe(
      dataframe= dfdata_test,
      batch_size=1,
      class_mode='categorical',
      target_size=(28, 28),
      x_col='filePath',
      seed=10,
    reshuffle_each_iteration=False
      y_col='shape',
      color_mode = 'grayscale')
```

```python

d_inv = {0:'diamond', 1:'hex', 2:'round', 3:'square',4:'triangle'}
plt.figure(figsize=(16,16))
batches  =1
for x,y in test_gen_shape_new:
        batches = batches + 1
        y_hat = best_aug.predict(x, verbose=0)
        x = np.squeeze(x)
        if batches < 20:
            plt.subplot(5,5,batches)
            plt.imshow(x)
            plt.title("GT-{}, Pred-{}".format(d_inv[np.argmax(y[0])], d_inv[np.argmax(y_hat[0])] ))
            
        else:
            break
        
        plt.axis('off')
```
