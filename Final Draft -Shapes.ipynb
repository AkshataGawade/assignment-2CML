# Data Loading for training model

Data is loaded in image generator and image is rescale between 0-1 to bring all the image in same range and save time to process image with high range.
<p>And 'color mode' is set as grayscale and 'class mode' as categorical since its classification problem.</p>
<p>Batch size = 32  standarized batch size.
This also means 32 images from the training dataset will be used to estimate the error gradient before the model weights are updated.</p>

```python
idg_train=ImageDataGenerator(rescale=1/255.0,data_format='channels_last')
idg_val=ImageDataGenerator(rescale=1/255.0,data_format='channels_last')
```

```python
train_gen=idg_train.flow_from_dataframe(dataframe=train_df,
                                        batch_size=32,
                                        seed=10,
                                        class_mode='categorical',
                                        target_size=(28, 28),
                                        x_col='Paths',
                                        y_col='Shapes',
                                        color_mode = 'grayscale')
val_gen=idg_val.flow_from_dataframe(dataframe=val_df,
                                       batch_size=32,
                                       seed=10,
                                       class_mode='categorical',
                                       target_size=(28, 28),
                                       x_col='Paths',
                                       y_col='Shapes',
                                       color_mode = 'grayscale'
                                      )
```

# Baseline Model
At first Tiny model is setup, here image data is falten and 32 neuron hidden layer with relu activation function is used.
Since in eda we saw there are 5 different type of shape last output layer is 5 label and activation function is softmax.
<p>Model is trained on 50 epochs and for model compilation following is used:</p>
- optimizer:: 'adam' gives faster computational time and yields better results.<br>
- loss:: 'categorical_crossentropy' since its multi class classification.<br>
- metrics:: 'categorical_accuracy' since its multi class classification.<br>

```python
baseline_model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28,1)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(5,activation='softmax')

])
```

```python
baseline_model.summary()
```

```python
baseline_model.compile(optimizer='adam',loss='categorical_crossentropy',
              metrics=['categorical_accuracy'])
```

```python
fitted_baseline=baseline_model.fit(train_gen,validation_data=val_gen,batch_size=32,epochs=50,shuffle=False)
```

```python
plt.title('shapes acc')
plt.plot(fitted_baseline.history['categorical_accuracy'],label='tr_acc')
plt.plot(fitted_baseline.history['val_categorical_accuracy'],label='val_acc')
plt.legend(loc='best')
```

```python
plt.plot(fitted_baseline.history['loss'],label='tr_loss')
plt.plot(fitted_baseline.history['val_loss'],label='val_loss')
plt.legend(loc='best')
```

### Observation
- Accuracy of training and validation set has increase with number of epochs.
- Loss function is going down with increase in epochs and accuracy of the model is increasing.
- There are some epochs where accuracy of validation set is greater than accuracy of the training set.
- Train and Test data are align till 35 epoch after that there is slight variation.
- Accuracy increases steeply till 5 epochs and than accuracy show stable increase

# Handling imbalance 
From the EDA we can see that there is high class imbalance accross the distribution of shapes:
<p>Using the weights of handle these imbalance, weights will increase pentalty of for loss function with class with less data is predicated wrong.<br>Thus ML model will train the class with less data fair chance.

```python
weights = compute_class_weight("balanced",np.unique(train_gen.classes),train_gen.classes)
```

```python
weights_dict={}
for i in range(weights.shape[0]):
    weights_dict[np.unique(train_gen.classes)[i]]=weights[i]
```

```python
weights_dict
```

# Training weighted model
### Observations
- Baseline model is kept same.
- Class weight is added while compiling the model.

```python
plt.title('shapes acc')
plt.plot(fitted_baseline_weighted.history['categorical_accuracy'],label='tr_acc')
plt.plot(fitted_baseline_weighted.history['val_categorical_accuracy'],label='val_acc')
plt.legend(loc='best')
```

# Creating image data generators with augmentation

Initializing ImageDataGenerator to make the data(Images), alterable to consider more test cases.
For example: If a triangle is upside down its still should be recognized as a triangle by the model.
<p>
Following image alteration were considered:<br>
    - rotation_range- If the test Image is rotated<br>
    - horizontal_flip/vertical_flip- To recognized upside down/flip images.<br>
    - zoom_range- To recognize zoomed images which may conatin same shape but of a different size.<br>
    - fill_mode- To higlight image features by extending nearby pixels.

```python
new_idg_train=ImageDataGenerator(rescale=1/255.0,
                                 data_format='channels_last',
                                 rotation_range=95, 
                                 zoom_range=0.2,
                                 fill_mode='nearest',
                                 horizontal_flip=True, 
                                 vertical_flip=True,)
new_idg_val=ImageDataGenerator(rescale=1/255.0,
                                 data_format='channels_last',
                                 rotation_range=95,
                                 zoom_range=0.2,
                                 fill_mode='nearest',
                                 horizontal_flip=True,
                                 vertical_flip=True)
new_train_gen=new_idg_train.flow_from_dataframe(dataframe=train_df,
                                        batch_size=32,
                                        seed=10,
                                        class_mode='categorical',
                                        target_size=(28, 28),
                                        x_col='Paths',
                                        y_col='Shapes',
                                        color_mode = 'grayscale')
new_val_gen=new_idg_val.flow_from_dataframe(dataframe=val_df,
                                       batch_size=32,
                                       seed=10,
                                       class_mode='categorical',
                                       target_size=(28, 28),
                                       x_col='Paths',
                                       y_col='Shapes',
                                       color_mode = 'grayscale'
                                      )
```

# Using conv2D and maxpooling 
- Conv2D is used to highlight image features such as edges.
- MaxPooling maximum values for each patch (2x2) of the image map.

```python
best_aug = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(5,(3,3),activation='relu',input_shape=(28,28,1)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    tf.keras.layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    tf.keras.layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    tf.keras.layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),

    tf.keras.layers.Dense(5,activation='softmax')
    
])
best_aug.compile(optimizer='adam',loss='categorical_crossentropy',
              metrics=['categorical_accuracy'])
best_aug_fit=best_aug.fit(new_train_gen,validation_data=new_val_gen,batch_size=32,epochs=100,class_weight=weights_dict,shuffle=False)
```

```python
plt.title('shapes acc')
plt.plot(best_aug_fit.history['categorical_accuracy'],label='tr_acc')
plt.plot(best_aug_fit.history['val_categorical_accuracy'],label='val_acc')
plt.legend(loc='best')
```

# Ultimate Judgement: Testing the best model with new data.

```python
path ="./Final_Testing/"
lst = []
for root, dirs, files in os.walk(path):
    for file in files:
        if(file.endswith(".png")):
            root1=root.split("/")[-1].split("\\")
            filepath=path+root1[0]+'/'+root1[1]+'/'+file
            lst.append([root1[0], root1[1], file, filepath])
dfdata_test = pd.DataFrame(lst, columns=['shape', 'sign', 'imageName','filePath'])
```

```python
new_idg_train_shape=ImageDataGenerator(rescale=1/255.0,
                                 data_format='channels_last',
                                 rotation_range=95, 
                                 zoom_range=0.2,
                                 fill_mode='nearest',
                                 horizontal_flip=True, 
                                 vertical_flip=True)

test_gen_shape_new = new_idg_train_shape.flow_from_dataframe(
      dataframe= dfdata_test,
      batch_size=1,
      class_mode='categorical',
      target_size=(28, 28),
      x_col='filePath',
      seed=10,
      y_col='shape',
      color_mode = 'grayscale')
```

```python

d_inv = {0:'diamond', 1:'hex', 2:'round', 3:'square',4:'triangle'}
plt.figure(figsize=(16,16))
batches  =1
for x,y in test_gen_shape_new:
        batches = batches + 1
        y_hat = best_aug.predict(x, verbose=0)
        x = np.squeeze(x)
        if batches < 20:
            plt.subplot(5,5,batches)
            plt.imshow(x)
            plt.title("GT-{}, Pred-{}".format(d_inv[np.argmax(y[0])], d_inv[np.argmax(y_hat[0])] ))
            
        else:
            break
        
        plt.axis('off')
```
